```python
demands_df = pd.read_csv('../Data/lebanon demand - Sheet1.csv')
demands_df.drop(columns=['x','Demands'],inplace=True)
```

```python
demand_dict = demands_df.to_dict('list')
```

```python
for k in demand_dict.keys():
    demand_dict[k] = [x for x in demand_dict[k] if x is not np.nan]
```

```python
df['demandtypeone'].value_counts()
```

```python
def update_demands(demand):
    
    for k in demand_dict.keys():    
        if demand in demand_dict[k]:
            return k
```

```python
df['demandtypeone'] = df.demandtypetwo.apply(lambda x: update_demands(x) if x != 'Other' else 'Unknown')
```
