```python
import pandas as pd
import nltk

#nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stopwords = set(stopwords.words('arabic'))
```

```python
df = pd.read_csv('../Data/Lebanon_89-2010 - Sheet1.csv', header=1)
df.dropna(subset=['narrative'], inplace=True)
```

```python
df['narrative_words'] = df['narrative'].apply(nltk.word_tokenize)
df['narrative_words'] = df['narrative_words'].apply(lambda x: [word.lower() for word in x if word.isalpha() and word not in stopwords])

all_words = [word for words in df['narrative_words'] for word in words]
fdist = nltk.FreqDist(all_words)
```
